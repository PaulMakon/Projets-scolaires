```python
import re
from datetime import datetime, timedelta
pattern = r'(\d{2}/\d{2}/\d{4}) (\d{2}:\d{2})'


def extract_all_datetime(text):
    matches = re.findall(pattern, text)
    times = [match[0] for match in matches if match[0]]
    dates = [match[1] for match in matches if match[1]]
    return list(zip(dates, times))[-1]


df["l'heure de sortir"] = df[" Localisation->Salles"].apply(extract_all_datetime)

df["l'heure de sortir"] = df["l'heure de sortir"].astype(str)
print(df["l'heure de sortir"])
```

```python
# code pour prédicition de la loi d'arrivée des patients 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split


df['date_time'] = pd.to_datetime(df[' Informations du passage->date et heure d\'arrivée'])

# Agréger les données par jour
df_daily = df.resample('D', on='date_time').size().reset_index()
df_daily.columns = ['date', 'arrivals']

# Diviser les données en ensembles d'entraînement et de test
train_size = int(len(df_daily) * 0.8)  # 80% des données pour l'entraînement
train, test = df_daily[:train_size], df_daily[train_size:]

# Visualiser les données agrégées
plt.figure(figsize=(10, 6))
plt.plot(train['date'], train['arrivals'], label='Train')
plt.plot(test['date'], test['arrivals'], label='Test')
plt.title('Nombre d\'arrivées de patients par jour')
plt.xlabel('Date')
plt.ylabel('Nombre d\'arrivées')
plt.legend()
plt.show()

# Décomposition saisonnière
decomposition = seasonal_decompose(train['arrivals'], model='additive', period=7)  # Période hebdomadaire
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Test de stationnarité
adf_result = adfuller(train['arrivals'])
print('Augmented Dickey-Fuller Test:')
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')

# Modèle ARIMA
model = ARIMA(train['arrivals'], order=(1,1,1))  # Modèle ARIMA(1,1,1)
results = model.fit()

# Prédictions
forecast_steps = len(test)  # Nombre de pas de temps pour les prédictions
forecast = results.forecast(steps=forecast_steps)

# Visualisation des prédictions
plt.figure(figsize=(10, 6))
plt.plot(train['date'], train['arrivals'], label='Train')
plt.plot(test['date'], test['arrivals'], label='Test')
plt.plot(test['date'], forecast, label='Prédictions')
plt.title('Prédictions de la loi d\'arrivée des patients')
plt.xlabel('Date')
plt.ylabel('Nombre d\'arrivées')
plt.legend()
plt.show()
```
